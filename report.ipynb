```python
# 初期設定
from Tree import BinaryTree,AVLTree
import random

random.seed(1234)
binarytree = BinaryTree()

arr_add = list(range(1,15))
arr_del = list(range(1,15))
random.shuffle(arr_add)
random.shuffle(arr_del)
```

```python
# 挿入処理
print('追加するサンプル : {}\n'.format(str(arr_add)))
print('追加する値, 現在のノード')

for e in arr_add:
    binarytree = binarytree.insert(v=e)
    print('{0:>8}, {1}'.format(e,str(binarytree.traverse_in_order())))
    
binarytree.export_graphviz_tree()
```

```python
# 検索処理
# 検索にHitすれば値の比較回数(計算量)を出力, Hitしなければ0を出力する
# 例 : 14 -> 3, 20 -> 0
binarytree.search(14),binarytree.search(20)
```

```python
# 削除処理
print('削除するサンプル : {}\n'.format(str(arr_del)))
print('追加する値, 現在のノード')

for e in arr_del:
    binarytree = binarytree.delete(e)
    print('{0:>8}, {1}'.format(e,str(binarytree.traverse_in_order())))
    
binarytree.export_graphviz_tree()
```

```python
# 最悪計算量の例
# 線形探索と同等のO(N)

for e in range(1,16):
    binarytree = binarytree.insert(v=e)
    
print(binarytree.search(14))
binarytree.export_graphviz_tree()
```

```python
# 初期設定
avltree = AVLTree()
```

```python
# 挿入処理
print('追加するサンプル : {}\n'.format(str(arr_add)))
print('追加する値, 現在のノード')

for e in arr_add:
    avltree = avltree.insert(v=e)
    print('{0:>8}, {1}'.format(e,str(avltree.traverse_in_order())))
    
avltree.export_graphviz_tree()
```

```python
# 削除処理
print('削除するサンプル : {}\n'.format(str(arr_del)))
print('追加する値, 現在のノード')

for e in arr_del:
    avltree = avltree.delete(e)
    print('{0:>8}, {1}'.format(e,str(avltree.traverse_in_order())))
    
avltree.export_graphviz_tree()
```

```python
# 二分木の場合の最悪計算量の例
# この例では計算量がlog(N-1)に減少

for e in range(15):
    avltree = avltree.insert(v=e)
    
print(avltree.search(14))
avltree.export_graphviz_tree()
```

```python
# AVL木の最悪計算量の例

avltree = AVLTree()

for e in arr_add:
    avltree = avltree.insert(v=e)

print(avltree.search(8))
avltree.export_graphviz_tree()
```
